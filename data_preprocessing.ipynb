```python
import pandas as pd
import numpy as np
from scipy.io import arff
```

## All the data

Training

```python
data, meta = arff.loadarff("data/kdd_diff/KDDTrain+.arff")
training_df = pd.DataFrame(data)
```

```python
for col in training_df.select_dtypes(include=['object']):
    training_df[col] = training_df[col].map(lambda x: x.decode('utf-8') if isinstance(x, bytes) else x)
```

```python
training_df
```

Test

```python
data, meta = arff.loadarff("data/kdd_diff/KDDTest+.arff")
test_df = pd.DataFrame(data)
```

```python
for col in test_df.select_dtypes(include=['object']):
    test_df[col] = test_df[col].map(lambda x: x.decode('utf-8') if isinstance(x, bytes) else x)
```

```python
test_df
```

Save to csv file

```python
training_df.to_csv("data/training_full.csv", index=False)
test_df.to_csv("data/test_full.csv", index=False)
```
